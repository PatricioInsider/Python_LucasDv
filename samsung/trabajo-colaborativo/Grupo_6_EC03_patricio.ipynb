Un grupo de estudiantes registró durante 30 días la **cantidad de horas que pasaron usando redes sociales**.

Su tarea será **analizar esta Serie** aplicando operaciones estadísticas y visualizaciones para entender mejor los hábitos de uso.

4. **Filtrado y condiciones**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Creando Series**

```python
horas = np.random.randint(0, 11, 30)
#print(horas)
dias = [f"Día {i+1}" for i in range(30)]

# Convertimos la lista en una Serie de Pandas (como una columna con nombre)
serie = pd.Series(horas, index=dias)
print(serie.to_frame())
```

**Explorando las series**

```python
serie.name = "Horas de uso de redes sociales"
print(serie.name)
print("Primeros 5 días:")
print(serie.head())
print("Últimos 5 días:")
print(serie.tail())

#Valor especifico
print("Día 7: ", serie.loc["Día 7"]) #acceso por etiqueta

# Ocupando atributos para ver la longitud tipo de datos y valores nulso.
print("Longitud:", serie.size)
print("Tipo de datos:", serie.dtype)
print("Valores nulos:", serie.isnull().sum())
```

**Estadísticas Básicas**

```python
# Calculando: promedio, mediana, mínimo, máximo, varianza y desviación estándar.
# Identificando los días con mayor y menor uso de redes sociales.

print("Promedio:", serie.mean())
print("Mediana:", serie.median())
print("Mínimo:", serie.min())
print("Máximo:", serie.max())
print("Varianza:", serie.var())
print("Desviación estándar:", serie.std())

print("Día de mayor uso:", serie.idxmax(), "con", serie.max(), "horas")
print("Día de menor uso:", serie.idxmin(), "con", serie.min(), "horas")
```

**Filtrado y condiciones**

```python
print("Dias con más de 6 horas")
print(serie[serie > 6])

print("Cantidad de dias con menos de 2 horas:", (serie < 2).sum())
```

**Visualizaciones**

```python
from os import set_inheritable
#Gráfico de línea — Evolución diaria

plt.figure(figsize=(12, 6))
serie.plot(kind='line', marker='o', color='red')
plt.title("Evolución diaria del uso de redes sociales")
plt.xlabel("Día")
plt.ylabel("Horas")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


#Grafico de hist

plt.figure(figsize=(12, 6))
serie.plot(kind='hist', bins=10, color='green', edgecolor='black')
plt.title("Distribución de horas de uso")
plt.xlabel("Horas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.tight_layout()
plt.show()

#grafico de baras
plt.figure(figsize=(12, 6))
serie.plot(kind='bar', color='yellow')
plt.title("Horas de uso por día")
plt.xlabel("Día")
plt.ylabel("Horas")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

#Grafico de pastel
plt.figure(figsize=(12, 6))
serie.plot(kind='pie', color='coral')
plt.title("Horas de uso por día")
plt.xlabel("Día")
plt.ylabel("Horas")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


```
