## Q 08-01

### 1. ¿Cuál es el problema que ocurre cuando el número de nodos es pequeño al analizar redes neuronales artificiales?
1. Si la cantidad de nodos es pequeña, la cantidad de operaciones a procesar es pequeña, por lo que se realiza rápidamente.
2. No se pueden crear límites complejos para la toma de decisiones al realizar un modelo de análisis.
3. No se puede realizar el algoritmo de retropropagación que ajusta pesos y umbrales.
4. La cantidad de nodos no tiene nada que ver con el modelo de análisis.

**Respuesta: 2**

Cuando el número de nodos es pequeño, la red neuronal tiene capacidad limitada para aprender patrones complejos. No puede crear fronteras de decisión complejas necesarias para clasificar datos no lineales, resultando en underfitting.

## Q 08-02

### 2. ¿Cuál es el concepto que se describe a continuación?
`Matriz multidimensional con 3 componentes: rango, forma, tipo`

**Respuesta: Tensor**

Un tensor es una estructura de datos multidimensional fundamental en TensorFlow. Sus 3 componentes son:
- **Rango (rank):** número de dimensiones
- **Forma (shape):** tamaño de cada dimensión
- **Tipo (dtype):** tipo de datos (float32, int32, etc.)

## Q 08-03

### ¿Qué sucede si el error de verificación aumenta constantemente cuando graficamos el error de verificación para cada época utilizando el método de descenso de pendiente por lotes? Además, ¿cómo se puede solucionar este problema?

**Respuesta:**

**Problema:** Si el error de validación aumenta constantemente, indica overfitting. El modelo memoriza los datos de entrenamiento pero no generaliza bien.

**Soluciones:**
1. Reducir la tasa de aprendizaje (learning rate)
2. Aplicar regularización (L1, L2, dropout)
3. Early stopping: detener el entrenamiento cuando el error de validación empieza a aumentar
4. Aumentar datos de entrenamiento
5. Reducir la complejidad del modelo

## Q 08-04

### La siguiente figura son datos con una estructura tridimensional.
1) Presente las ecuaciones de convolución para (a)
2) Presente las ecuaciones de convolución para (b)

**Respuesta:**

Basándome en la imagen mostrada:

**1) Ecuación de convolución para (a):**

$$y_{ij} = \sum_{m=0}^{2} \sum_{n=0}^{2} w_{mn} \cdot x_{(i+m)(j+n)} + b$$

Donde:
- $w_{mn}$ es el kernel/filtro 3×3
- $x_{(i+m)(j+n)}$ es la entrada
- $b$ es el bias

**2) Ecuación de convolución para (b) - con múltiples canales:**

$$y_{ij}^{k} = \sum_{c=0}^{C-1} \sum_{m=0}^{2} \sum_{n=0}^{2} w_{mn}^{ck} \cdot x_{(i+m)(j+n)}^{c} + b^{k}$$

Donde:
- $C$ es el número de canales de entrada
- $k$ es el índice del filtro de salida
- $w_{mn}^{ck}$ es el peso del filtro para el canal $c$

## Q 08-05

### Aplique la función softmax cuando la salida de la red neuronal sea $(0.4,2.0,0.001,0.32)^T$ y escriba el resultado.

**Respuesta:**

La función softmax es: $\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$

Para $(0.4, 2.0, 0.001, 0.32)^T$:

$$e^{0.4} = 1.4918$$
$$e^{2.0} = 7.3891$$
$$e^{0.001} = 1.0010$$
$$e^{0.32} = 1.3771$$

$$\sum e^{x_i} = 11.2590$$

**Resultado:**
$$\text{softmax} = \begin{pmatrix} 0.1325 \\ 0.6563 \\ 0.0889 \\ 0.1223 \end{pmatrix}$$

O aproximadamente: **(0.13, 0.66, 0.09, 0.12)**
