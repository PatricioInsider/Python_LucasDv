## Q 07-01

### ¿Cuál de las siguientes afirmaciones sobre la minería de textos es incorrecta?
1. Puedes conocer la reacción ante un grupo específico.
2. Se puede obtener información sobre oportunidades de ventas.
3. Se pueden establecer estrategias competitivas monitoreando otras marcas.
4. El mismo método se puede aplicar a varios idiomas.

**Respuesta: 4**

La afirmación incorrecta es la opción 4. El mismo método NO se puede aplicar directamente a varios idiomas, ya que cada idioma tiene características lingüísticas únicas (gramática, morfología, sintaxis) que requieren técnicas de preprocesamiento y modelos específicos. Por ejemplo, el tokenizado, stemming y stopwords son diferentes para cada idioma.

## Q 07-02

### ¿Cuál de las siguientes afirmaciones no es cierta sobre el uso de las redes sociales?
1. Es posible saber de cuántos grupos está compuesta la red.
2. Es posible conocer clientes influyentes.
3. Es posible ver el cambio a lo largo del tiempo.
4. Es posible saber si los clientes se irán la próxima vez.

**Respuesta: 4**

La afirmación no cierta es la opción 4. Aunque el análisis de redes sociales puede identificar patrones de comportamiento y predecir tendencias, NO puede determinar con certeza si un cliente específico se irá la próxima vez. Esto requeriría modelos predictivos adicionales de churn (abandono) que van más allá del simple análisis de redes sociales.

## Q 07-03

### ¿Qué significa el término separar una raíz de una palabra cuya forma ha sido modificada para extraerla? ¿Cuál es la palabra que es objeto de análisis de morfemas?

**Respuesta:**

El término es Stemming (o Lematización si se busca la forma canónica).

- **Stemming**: Proceso de reducir palabras a su raíz o stem eliminando sufijos y prefijos (ej: "corriendo" → "corr").
- **Lemmatization**: Proceso más sofisticado que reduce palabras a su forma base o lema usando análisis morfológico (ej: "corriendo" → "correr").

La palabra objeto de análisis de morfemas es el morfema, que es la unidad mínima con significado en una lengua.

## Q 07-04

### ¿Qué significa el término un conjunto de materiales que pueden mostrar los aspectos esenciales del lenguaje como material de investigación requerido en cada campo de la investigación del lenguaje?

**Respuesta:**

El término es **Corpus** (plural: corpora).

Un corpus es una colección estructurada y representativa de textos en formato digital que se utiliza para análisis lingüístico, entrenamiento de modelos de NLP y estudios del lenguaje. Ejemplos: corpus de noticias, corpus de redes sociales, corpus médico, etc.

## Q 07-05

### Explique sobre TF (Frecuencia de términos) y escriba cómo calcular TF (fórmula).

**Respuesta:**

**TF (Term Frequency)** mide la frecuencia con que aparece un término en un documento. Indica la importancia de una palabra dentro de un documento específico.

**Fórmula:**

$$TF(t, d) = \frac{\text{Número de veces que el término } t \text{ aparece en el documento } d}{\text{Número total de términos en el documento } d}$$

O de forma simplificada:

$$TF(t, d) = \frac{f_{t,d}}{\sum_{t' \in d} f_{t',d}}$$

Donde:
- $f_{t,d}$ = frecuencia del término $t$ en el documento $d$
- El denominador es el total de términos en el documento

**Ejemplo:** Si la palabra "inteligencia" aparece 5 veces en un documento de 100 palabras, entonces TF = 5/100 = 0.05
