## Ejercicio 2: DBSCAN agrupación.

DBSCAN (agrupación espacial basada en la densidad de aplicaciones con ruido)
1. Explique cómo funciona el algoritmo DBSCAN para generar agrupaciones.
2. Enumere algunas de las ventajas de este algoritmo.
3. Enumere algunos ejemplos donde se podria aplicar.



DBSCAN es un algoritmo de clustering basado en **densidad** que agrupa puntos que están cerca unos de otros y marca como **outliers** (ruido) los puntos que están aislados.
#### **Tipos de puntos:**

- **Core point (punto núcleo):** Tiene al menos MinPts puntos dentro de su radio ε (incluyéndose a sí mismo)
- **Border point (punto frontera):** Está en el radio ε de un core point, pero no tiene suficientes vecinos para ser core
- **Noise point (ruido):** No es ni core ni border - está aislado

#### **Proceso del algoritmo:**

```
1. Para cada punto no visitado:
   a. Marcarlo como visitado
   b. Encontrar todos los puntos vecinos dentro del radio ε
   
2. Si el punto tiene menos de MinPts vecinos:
   - Marcarlo temporalmente como ruido
   
3. Si el punto tiene MinPts o más vecinos:
   - Crear un nuevo cluster
   - Agregar el punto al cluster
   - Expandir el cluster recursivamente:
     * Para cada vecino no visitado:
       - Marcarlo como visitado
       - Si también es core point, agregar sus vecinos a la lista
       - Agregarlo al cluster actual
```

**Visualización conceptual:**
```
    ●  ●  ●          ●
  ●  ●  ●  ●      ●  ●
    ●  ●  ●          ●
                        ○  (ruido)
  Cluster 1      Cluster 2
```


### 2. Ventajas del algoritmo DBSCAN

- A diferencia de K-Means, no necesitas saber cuántos grupos hay

- K-Means solo encuentra clusters esféricos
 - DBSCAN puede detectar formas irregulares, curvas, anillos, etc.


- Los identifica como ruido en lugar de forzarlos a pertenecer a un cluster
- K-Means es muy sensible a outliers

 - Puede manejar clusters de diferentes tamaños y densidades
 - Siempre produce los mismos resultados (a diferencia de K-Means que depende de inicialización aleatoria)


````python
# Ejemplo práctico de DBSCAN
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons, make_blobs

# Crear datos con formas no esféricas (dos lunas)
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
clusters = dbscan.fit_predict(X)

# Visualizar resultados
plt.figure(figsize=(10, 5))

# Gráfico de los clusters encontrados
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50)
plt.title('DBSCAN - Clusters encontrados')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Identificar ruido
plt.subplot(1, 2, 2)
noise = clusters == -1
plt.scatter(X[~noise, 0], X[~noise, 1], c=clusters[~noise], cmap='viridis', s=50, label='Clusters')
plt.scatter(X[noise, 0], X[noise, 1], c='red', s=50, marker='x', label='Ruido')
plt.title('DBSCAN - Ruido identificado')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Número de clusters encontrados: {len(set(clusters)) - (1 if -1 in clusters else 0)}")
print(f"Número de puntos de ruido: {list(clusters).count(-1)}")
````

## Apply NLP for Text Cleaning

### Subtask:
Use Natural Language Processing (NLP) techniques to eliminate unnecessary words from the relevant text fields in the dataset.


````python
import re
import nltk
from nltk.corpus import stopwords

# Descargar stopwords de NLTK si aún no han sido descargadas
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

# Identificar columnas de texto relevantes para la limpieza NLP
text_columns = ['Title', 'Description', 'Genres', 'Cast', 'Director', 'Production Country']

# Rellenar valores NaN en las columnas de texto relevantes con cadenas vacías para evitar errores durante el procesamiento de texto
for col in text_columns:
    df_netflix[col] = df_netflix[col].fillna('')

# Función para limpiar texto: convertir a minúsculas, eliminar puntuación, caracteres especiales y números
def clean_text(text):
    text = text.lower()  # Convertir a minúsculas
    text = re.sub(r'[^a-z\s]', '', text)  # Eliminar puntuación, caracteres especiales y números
    text = re.sub(r'\s+', ' ', text).strip() # Eliminar espacios extra
    return text

# Función para eliminar stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

# Aplicar limpieza y eliminación de stopwords a las columnas de texto elegidas
for col in text_columns:
    df_netflix[f'{col}_cleaned'] = df_netflix[col].apply(clean_text)
    df_netflix[f'{col}_cleaned'] = df_netflix[f'{col}_cleaned'].apply(remove_stopwords)

print("Limpieza de texto y eliminación de stopwords aplicadas a las columnas relevantes.")
print("Primeras 5 filas del DataFrame con las nuevas columnas limpias:")
print(df_netflix[[col for col in text_columns] + [f'{col}_cleaned' for col in text_columns]].head())
````

### 6. Cosine Similarity - Explicación y Aplicación

#### **¿Qué es Cosine Similarity?**

La **similitud del coseno** mide el **ángulo** entre dos vectores en un espacio multidimensional. No mide la distancia, sino la **orientación**.

#### **Fórmula matemática:**

$$
\text{cosine similarity} = \frac{A \cdot B}{\|A\| \|B\|} = \frac{\sum_{i=1}^{n} A_i B_i}{\sqrt{\sum_{i=1}^{n} A_i^2} \sqrt{\sum_{i=1}^{n} B_i^2}}
$$

Donde:
- **A · B** = Producto punto de los vectores
- **||A||** y **||B||** = Magnitudes (normas) de los vectores

#### **Interpretación:**
- **1.0** = Vectores idénticos (ángulo de 0°)
- **0.0** = Vectores perpendiculares (ángulo de 90°, sin relación)
- **-1.0** = Vectores opuestos (ángulo de 180°)

#### **¿Por qué es útil para recomendaciones?**

1. **Independiente de la magnitud:** Solo importa la dirección, no el tamaño
   - Ejemplo: Un documento largo y uno corto sobre el mismo tema tendrán alta similitud

2. **Eficiente computacionalmente:** Cálculo rápido incluso con muchas dimensiones

3. **Funciona bien con TF-IDF:** Ideal para comparar documentos de texto

#### **Ejemplo visual:**
```
Documento A: [3, 4]  (habla mucho de "acción" y "aventura")
Documento B: [6, 8]  (también habla de "acción" y "aventura", pero más extenso)
Documento C: [4, 1]  (habla más de "acción" que de "aventura")

Cosine(A, B) = 1.0  (misma dirección, muy similares)
Cosine(A, C) = 0.88 (dirección parecida, algo similares)
```

#### **Proceso en nuestro sistema:**
1. Convertir texto a vectores numéricos (TF-IDF)
2. Calcular cosine similarity entre todos los pares
3. Para una película dada, encontrar las más similares

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df_netflix['combined_features'] = df_netflix['Title_cleaned'] + ' ' + \
                                  df_netflix['Description_cleaned'] + ' ' + \
                                  df_netflix['Genres_cleaned'] + ' ' + \
                                  df_netflix['Cast_cleaned'] + ' ' + \
                                  df_netflix['Director_cleaned'] + ' ' + \
                                  df_netflix['Production Country_cleaned']

df_netflix['combined_features'] = df_netflix['combined_features'].fillna('')

print("Combined features created successfully.")

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df_netflix['combined_features'])

print("TF-IDF matrix created successfully. Shape:", tfidf_matrix.shape)

cosine_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

print("Cosine similarity matrix created successfully. Shape:", cosine_sim_matrix.shape)


indices = pd.Series(df_netflix.index, index=df_netflix['Title']).drop_duplicates()

print("Movie title to index mapping created successfully.")
````

````python
def get_recommendations(title, cosine_sim=cosine_sim_matrix, df=df_netflix, indices=indices):
    # Obtener el índice de la película que coincide con el título
    if title not in indices.index:
        print(f"Movie '{title}' not found in the database.")
        return []

    idx = indices[title]

    # Obtener las puntuaciones de similitud para todas las películas con esa película
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Ordenar las películas basándose en las puntuaciones de similitud en orden descendente
    # Excluir la película en sí (puntuación de 1) de las recomendaciones
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Obtener las 10 películas más similares (excluyendo la propia)

    # Obtener los índices de las películas
    movie_indices = [i[0] for i in sim_scores]

    # Devolver los títulos de las 10 películas más similares
    return df['Title'].iloc[movie_indices].tolist()

print("Recommendation function 'get_recommendations' defined successfully.")
````
