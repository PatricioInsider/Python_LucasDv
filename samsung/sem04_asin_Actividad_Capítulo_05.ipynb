# Actividad: Ordenamiento de Burbuja


## Instrucciones

1. Explique brevemente y en forma clara cómo funciona el algoritmo de ordenamiento de burbuja.

2. Desarrollo un función donde se use el ordenamiento de burbuja aplicada a una lista de valores enteros.



```python
# Espacio para desarrollar el código requerido, recuerde agregar algunos

# comentarios breves sobre su funcionamiento.
# El algoritmo de burbuja compara elementos adyacentes de una lista y los intercambia si están en el orden incorrecto.
# Este proceso se repite varias veces hasta que la lista está completamente ordenada.
# Es sencillo pero no muy eficiente para listas grandes.

def bubble_sort(lista):

    n = len(lista)
    for i in range(n):
        # En cada pasada, el elemento más grande 'burbujea' al final de la lista no ordenada
        for j in range(0, n-i-1):
            if lista[j] > lista[j+1]:
                # Intercambia si el elemento actual es mayor que el siguiente
                lista[j], lista[j+1] = lista[j+1], lista[j]
    return lista

valores = [5, 2, 9, 1, 5, 6]
print("Lista original:", valores)
print("Lista ordenada:", bubble_sort(valores.copy()))




```

# Actividad: Función Swap.

1. Escriba una función que permita el intercambion de elementos en una lista.
2. Los argumentos de esta función serán, la lista y los indices de los elementos que se desean intercambiar.
3. En la siguiente celda desarrollo el código requerido, recuerde agregar comentarias breves que expliquen su funcionamiento.


```python
# Función Swap: intercambia dos elementos de una lista dados sus índices.
def swap(lista, i, j):

    lista[i], lista[j] = lista[j], lista[i]

# Ejemplo de uso:
ejemplo = [10, 20, 30, 40]
print("Antes del swap:", ejemplo)
swap(ejemplo, 1, 3)
print("Después del swap:", ejemplo)
```

```python
# Celda para desarrollar 


# Quick Sort es un algoritmo de ordenamiento eficiente que utiliza el método de divide y vencerás.
# Selecciona un elemento como pivote y divide la lista en dos sublistas:
# una con elementos menores al pivote y otra con elementos mayores.
# Luego ordena recursivamente las sublistas.

def quick_sort(lista):
    if len(lista) <= 1:
        return lista
    else:
        pivote = lista[0]
        menores = [x for x in lista[1:] if x <= pivote]
        mayores = [x for x in lista[1:] if x > pivote]
        return quick_sort(menores) + [pivote] + quick_sort(mayores)

# Ejemplo de uso:
valores = [8, 3, 1, 7, 0, 10, 2]
print("Lista original:", valores)
print("Lista ordenada:", quick_sort(valores))




```

# Actividad: Quick Sort usando un valor de pivote aleatorio.

1. Explique brevemente el funcionamiento del algoritmo Quick Sort en la modaladidad de usar un valor de pivote aleatorio.
2. Comente las ventajas de esta variación.
3. Desarrollo el código necesarion para este algoritmo.
4. Agregue comentarios sobre su funcionamiento.



```python
# Espacio para desarrollar el código siguiendo las instrucciones.

import random

# En esta variante, el pivote se elige aleatoriamente, lo que ayuda a evitar el peor caso
# cuando la lista ya está casi ordenada o tiene muchos elementos repetidos.
# Ventajas: reduce la probabilidad de que el algoritmo tenga un rendimiento deficiente.

def quick_sort_random(lista):

    if len(lista) <= 1:
        return lista
    else:
        pivote_idx = random.randint(0, len(lista) - 1)
        pivote = lista[pivote_idx]
        menores = [x for i, x in enumerate(lista) if x <= pivote and i != pivote_idx]
        mayores = [x for i, x in enumerate(lista) if x > pivote and i != pivote_idx]
        return quick_sort_random(menores) + [pivote] + quick_sort_random(mayores)

valores = [8, 3, 1, 7, 0, 10, 2]
print("Lista original:", valores)
print("Lista ordenada (pivote aleatorio):", quick_sort_random(valores))
```
